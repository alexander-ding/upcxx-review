```python
from pathlib import Path
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
with open('test_result_real.json') as f:
    jsons = json.load(f)
xs = list(range(2, 72, 4))

data = {
    'pagerank': {},
    'bfs': {},
    'bf': {},
    'cc': {}
}

def get_data(jsons, xs, test, graph):
    def get(l, name):
        r = [d['time'] for d in l if not d['error'] and d['name'] == name]
        if len(r) > 0:
            return r[0]
        else:
            return -1
    xs_ret = {
        'upcxx': [],
        'mp': []
    }
    ys_ret = {
        'upcxx': [],
        'mp': []
    }
    for x in xs:
        result = jsons[str(x)][test]
        for kind in ['upcxx', 'mp']:
            r = get(result[kind], graph)
            if r == -1:
                continue
            xs_ret[kind].append(x)
            ys_ret[kind].append(r)
    return xs_ret, ys_ret

tests = ['pagerank', 'bfs', 'bf', 'cc']
graphs = ['facebook_combined', 'twitter_combined', 'gplus_combined', 'com-orkut.ungraph', 'com-youtube.ungraph']

data = {}
for test in tests:
    data[test] = {}
    for graph in graphs:
        xs_ret, ys_ret = get_data(jsons, xs, test, graph)
        data[test][graph] = (xs_ret, ys_ret)
```

```python
def graph_data(data, test, graph, **params):
    xs, ys = data[test][graph]
    
    fig, ax = plt.subplots(figsize=(12, 12), dpi=300)
    title = "{}: {}".format(test, graph)
    ax.set_title(title)
    for kind in ['upcxx', 'mp']:
        ax.plot(xs[kind], ys[kind], label=kind, marker='.')
        
    ax.legend()
    ax.set_xlabel("Nodes")
    ax.set_ylabel("Time (ms)")
    
    #outpath = output_dir / test / "{}_{}_{}.png".format(params['node_size'], params['edge_size'], params['neg_rate'])
    #plt.savefig(outpath, dpi=300)
    #plt.close(fig)
```

```python
graph_data(data, 'bfs', 'com-orkut.ungraph')
```
